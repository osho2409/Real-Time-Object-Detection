```python
import os

folders = [
    "/content/tvremote/train/labels",
    "/content/tvremote/valid/labels",
    "/content/tvremote/test/labels"
]

for folder in folders:
    for file in os.listdir(folder):

        if not file.endswith(".txt"):
            continue

        path = os.path.join(folder, file)

        new_lines = []

        with open(path, "r") as f:
            lines = f.readlines()

        for line in lines:

            vals = line.strip().split()

            if len(vals) == 5:
                # already detection format
                vals[0] = "2"     # remote class
                new_lines.append(" ".join(vals))

            else:
                cls = "2"

                coords = list(map(float, vals[1:]))

                xs = coords[0::2]
                ys = coords[1::2]

                xmin = min(xs)
                xmax = max(xs)

                ymin = min(ys)
                ymax = max(ys)

                x_center = (xmin + xmax) / 2
                y_center = (ymin + ymax) / 2

                width = xmax - xmin
                height = ymax - ymin

                new_lines.append(
                    f"{cls} {x_center} {y_center} {width} {height}"
                )

        with open(path, "w") as f:
            f.write("\n".join(new_lines))
```

DATA PREPROCESSING BECAUSE OPEN DATASET HAD MULTIPLE CLASSES

```python
import os

mapping = {
    0: 1,  # Cup -> cup
    1: 1,  # Cups -> cup
    2: 0,  # aqua -> bottle
    3: 0,  # bottle -> bottle
    4: 1   # cup -> cup
}

for split in ["train", "valid", "test"]:
    label_dir = f"/content/mugs_bottles/{split}/labels"

    for file in os.listdir(label_dir):
        if not file.endswith(".txt"):
            continue

        path = os.path.join(label_dir, file)

        new_lines = []

        with open(path, "r") as f:
            for line in f:
                parts = line.strip().split()

                cls = int(parts[0])

                parts[0] = str(mapping[cls])

                new_lines.append(" ".join(parts))

        with open(path, "w") as f:
            f.write("\n".join(new_lines))
```

```python
import os
import shutil

sources = [
    "/content/mugs_bottles",
    "/content/tvremote"
]

dest = "/content/final_dataset"

for source in sources:
    for split in ["train", "valid", "test"]:

        src_img = f"{source}/{split}/images"
        src_lbl = f"{source}/{split}/labels"

        dst_img = f"{dest}/{split}/images"
        dst_lbl = f"{dest}/{split}/labels"

        for file in os.listdir(src_img):
            shutil.copy(
                os.path.join(src_img, file),
                os.path.join(dst_img, file)
            )

        for file in os.listdir(src_lbl):
            shutil.copy(
                os.path.join(src_lbl, file),
                os.path.join(dst_lbl, file)
            )

print("Merge complete")
```
